```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("input/Flower.jpg")

if image is None:
    print("Image not found!")
else:
    print("Image loaded successfully.")
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
# Task 1
```

```python
# Read Image and Display Dimensions, Channels and File Size
```

```python
height, width, channels = image.shape

print("Height :", height)
print("Width :", width)
print("Channels :", channels)

file_size = os.path.getsize("input/Flower.jpg")

print("File Size :", file_size, "bytes")
```

```python
# Task 2
```

```python
# Convert Color Image to Grayscale
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()

cv2.imwrite("output/grayscale.jpg", gray)
```

```python
# Task 3
```

```python
#Resize Image
```

```python
# Small
small = cv2.resize(image, (300,200))

plt.imshow(cv2.cvtColor(small, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/resized_small.jpg", small)
```

```python
# medium
medium = cv2.resize(image, (600,400))

plt.imshow(cv2.cvtColor(medium, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/resized_medium.jpg", medium)
```

```python
# large
large = cv2.resize(image, (900,600))

plt.imshow(cv2.cvtColor(large, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/resized_large.jpg", large)
```

```python
# Task 4
```

```python
# Crop Different Regions
```

```python
# Center Crop 
crop1 = image[100:400,100:400]

plt.imshow(cv2.cvtColor(crop1, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/crop1.jpg", crop1)
```

```python
# Another Crop
crop2 = image[50:250,250:500]

plt.imshow(cv2.cvtColor(crop2, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/crop2.jpg", crop2)
```

```python
# Task 5
```

```python
# Rotate Image
```

```python
# 90°
rotate90 = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(cv2.cvtColor(rotate90, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/rotate90.jpg", rotate90)
```

```python
# 180°
rotate180 = cv2.rotate(image, cv2.ROTATE_180)

plt.imshow(cv2.cvtColor(rotate180, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/rotate180.jpg", rotate180)
```

```python
# 270°
rotate270 = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.imshow(cv2.cvtColor(rotate270, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/rotate270.jpg", rotate270)
```

```python
# Task 6
```

```python
#Flip Image
```

```python
# Horizontal
horizontal = cv2.flip(image,1)

plt.imshow(cv2.cvtColor(horizontal, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/horizontal.jpg", horizontal)
```

```python
# Vertical
vertical = cv2.flip(image,0)

plt.imshow(cv2.cvtColor(vertical, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/vertical.jpg", vertical)
```

```python
# Task 7
```

```python
# Draw Rectangle
rectangle = image.copy()

cv2.rectangle(rectangle,(50,50),(250,250),(0,255,0),3)

plt.imshow(cv2.cvtColor(rectangle,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/rectangle.jpg", rectangle)
```

```python
# Draw Circle
circle = image.copy()

cv2.circle(circle,(300,200),80,(255,0,0),3)

plt.imshow(cv2.cvtColor(circle,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/circle.jpg", circle)
```

```python
# Draw Line
line = image.copy()

cv2.line(line,(0,0),(500,500),(0,0,255),4)

plt.imshow(cv2.cvtColor(line,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/line.jpg", line)
```

```python
# Draw Polygon
polygon = image.copy()

points = np.array([[100,100],[300,80],[400,250],[200,350]], np.int32)

cv2.polylines(polygon,[points],True,(255,255,0),3)

plt.imshow(cv2.cvtColor(polygon,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/polygon.jpg", polygon)
```

```python
#Add Name and Today's Date
```

```python
text = image.copy()

cv2.putText(
    text,
    "Eman Fatima",
    (30,40),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (255,0,0),
    2
)

cv2.putText(
    text,
    "15-07-2026",
    (30,80),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (0,255,0),
    2
)

plt.imshow(cv2.cvtColor(text,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

cv2.imwrite("output/text.jpg", text)
```
