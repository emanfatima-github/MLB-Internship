```python
# import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Create output folder
os.makedirs("output", exist_ok=True)

# Load image
image_name = input("Enter image name: ")
image = cv2.imread("input/" + image_name)

if image is None:
    print("Image not found!")
else:
    processed = image.copy()

    while True:
        print("\n========== IMAGE PROCESSING TOOLKIT ==========")
        print("1. Convert to Grayscale")
        print("2. Resize Image")
        print("3. Rotate Image")
        print("4. Flip Image")
        print("5. Crop Image")
        print("6. Draw Shapes")
        print("7. Add Custom Text")
        print("8. Save Processed Image")
        print("9. Adjust Brightness and Contrast")
        print("10. Compare BGR and RGB")
        print("11. Display Original and Processed Images Side by Side")
        print("12. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            processed = cv2.cvtColor(processed, cv2.COLOR_BGR2GRAY)

            plt.imshow(processed, cmap="gray")
            plt.axis("off")
            plt.show()

        elif choice == "2":
            width = int(input("Enter width: "))
            height = int(input("Enter height: "))

            processed = cv2.resize(processed, (width, height))

            if len(processed.shape) == 2:
                plt.imshow(processed, cmap="gray")
            else:
                plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))

            plt.axis("off")
            plt.show()
            
        elif choice == "3":
            print("1. 90°")
            print("2. 180°")
            print("3. 270°")

            r = input("Choose rotation: ")

            if r == "1":
                processed = cv2.rotate(processed, cv2.ROTATE_90_CLOCKWISE)
            elif r == "2":
                processed = cv2.rotate(processed, cv2.ROTATE_180)
            elif r == "3":
                processed = cv2.rotate(processed, cv2.ROTATE_90_COUNTERCLOCKWISE)
            if len(processed.shape) == 2:
                plt.imshow(processed, cmap="gray")
            else:
                plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))

            plt.axis("off")
            plt.show()
            
        elif choice == "4":
            print("1. Horizontal")
            print("2. Vertical")

            f = input("Choose flip: ")

            if f == "1":
                processed = cv2.flip(processed, 1)
            elif f == "2":
                processed = cv2.flip(processed, 0)
            if len(processed.shape) == 2:
                plt.imshow(processed, cmap="gray")
            else:
                plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))

            plt.axis("off")
            plt.show()

        elif choice == "5":
            y1 = int(input("Start Row: "))
            y2 = int(input("End Row: "))
            x1 = int(input("Start Column: "))
            x2 = int(input("End Column: "))

            processed = processed[y1:y2, x1:x2]

            if len(processed.shape) == 2:
                plt.imshow(processed, cmap="gray")
            else:
                plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))

            plt.axis("off")
            plt.show()

        elif choice == "6":
            if len(processed.shape) == 2:
                processed = cv2.cvtColor(processed, cv2.COLOR_GRAY2BGR)

            cv2.rectangle(processed, (40,40), (200,200), (0,255,0), 3)
            cv2.circle(processed, (300,150), 60, (255,0,0), 3)
            cv2.line(processed, (0,0), (400,300), (0,0,255), 3)

            pts = np.array([[300,300],[350,350],[450,300],[420,220]], np.int32)

            cv2.polylines(processed, [pts], True, (255,255,0), 3)

            plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))
            plt.axis("off")
            plt.show()

        elif choice == "7":
            if len(processed.shape) == 2:
                processed = cv2.cvtColor(processed, cv2.COLOR_GRAY2BGR)

            name = input("Enter your name: ")

            cv2.putText(processed,
                        name,
                        (20,40),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1,
                        (255,0,255),
                        2)

            cv2.putText(processed,
                        "15-07-2026",
                        (20,80),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1,
                        (0,255,255),
                        2)

            plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))
            plt.axis("off")
            plt.show()

        elif choice == "8":
            cv2.imwrite("output/processed_image.jpg", processed)
            print("Image saved inside output folder.")

        elif choice == "9":
            brightness = int(input("Enter brightness (-100 to 100): "))
            contrast = float(input("Enter contrast (e.g. 1.0, 1.5, 2.0): "))

            processed = cv2.convertScaleAbs(processed, alpha=contrast, beta=brightness)

            if len(processed.shape) == 2:
                plt.imshow(processed, cmap="gray")
            else:
                plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))

            plt.axis("off")
            plt.show()

            cv2.imwrite("output/brightness_contrast.jpg", processed)

        elif choice == "10":
            rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            plt.figure(figsize=(10,5))

            plt.subplot(1,2,1)
            plt.imshow(image)
            plt.title("BGR (Wrong Colors)")
            plt.axis("off")

            plt.subplot(1,2,2)
            plt.imshow(rgb)
            plt.title("RGB (Correct Colors)")
            plt.axis("off")

            plt.show()
            cv2.imwrite("output/rgb_image.jpg", cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR))

        elif choice == "11":

            plt.figure(figsize=(12,6))

            plt.subplot(1,2,1)

            if len(image.shape) == 2:
                plt.imshow(image, cmap="gray")
            else:
                plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

            plt.title("Original Image")
            plt.axis("off")

            plt.subplot(1,2,2)

            if len(processed.shape) == 2:
                plt.imshow(processed, cmap="gray")
            else:
                plt.imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))

            plt.title("Processed Image")
            plt.axis("off")

            plt.show()

        elif choice == "12":
            print("Program Ended.")
            break            
```
