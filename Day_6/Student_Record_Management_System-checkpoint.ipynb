```python
# student_record_system.py
import json
import os
```

```python
FILE = "students.json"

# Load & Save
def load():
    if os.path.exists(FILE):
        with open(FILE, "r") as f:
            return json.load(f)
    return []

def save(records):
    with open(FILE, "w") as f:
        json.dump(records, f, indent=4)

# Operations
def add(records):
    name = input("Name: ").strip()
    if not name:
        print("Name can't be empty.")
        return
    try:
        age = int(input("Age: "))
    except ValueError:
        print("Age must be a number.")
        return
    grade = input("Grade: ").strip().upper()

    records.append({"name": name, "age": age, "grade": grade})
    save(records)
    print(f" {name} added!")

def view(records):
    if not records:
        print("No students yet.")
        return
    print(f"\n{'#':<4} {'Name':<20} {'Age':<6} {'Grade'}")
    print("-" * 38)
    for i, s in enumerate(records, 1):
        print(f"{i:<4} {s['name']:<20} {s['age']:<6} {s['grade']}")

def search(records):
    name = input("Search name: ").strip().lower()
    found = [s for s in records if name in s["name"].lower()]
    if found:
        for s in found:
            print(f"  → {s['name']}, Age: {s['age']}, Grade: {s['grade']}")
    else:
        print("Not found.")

def update(records):
    name = input("Name to update: ").strip().lower()
    for s in records:
        if s["name"].lower() == name:
            print(f"Current: {s}")
            new_name  = input(f"New name  [{s['name']}]: ").strip()
            new_age   = input(f"New age   [{s['age']}]: ").strip()
            new_grade = input(f"New grade [{s['grade']}]: ").strip()
            if new_name:  s["name"]  = new_name
            if new_grade: s["grade"] = new_grade.upper()
            if new_age:
                try:
                    s["age"] = int(new_age)
                except ValueError:
                    print("Invalid age — skipped.")
            save(records)
            print("Updated!")
            return
    print("Student not found.")

def delete(records):
    name = input("Name to delete: ").strip().lower()
    for i, s in enumerate(records):
        if s["name"].lower() == name:
            confirm = input(f"Delete '{s['name']}'? (yes/no): ")
            if confirm.lower() == "yes":
                records.pop(i)
                save(records)
                print("Deleted!")
            else:
                print("Cancelled.")
            return
    print("Student not found.")

# Main Menu
def main():
    records = load()
    print(f"\n {len(records)} student(s) loaded.")

    menu = {
        "1": ("Add student",    lambda: add(records)),
        "2": ("View all",       lambda: view(records)),
        "3": ("Search",         lambda: search(records)),
        "4": ("Update",         lambda: update(records)),
        "5": ("Delete",         lambda: delete(records)),
        "6": ("Exit",           None),
    }

    while True:
        print("\n──── MENU ────")
        for key, (label, _) in menu.items():
            print(f"  {key}. {label}")

        choice = input("\nChoice: ").strip()

        if choice == "6":
            print("Bye! Data saved.")
            break
        elif choice in menu:
            menu[choice][1]()
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```
