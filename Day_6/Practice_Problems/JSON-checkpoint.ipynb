```python
#Store student information in a JSON file.

import json

students = [
    {"name": "Bilal", "age": 20, "grade": "A"},
    {"name": "Zain", "age": 21, "grade": "A+"},
    {"name": "Usman", "age": 22, "grade": "B"},
]

with open("students.json", "w") as file:
    json.dump(students, file, indent = 4)

print("Student data saved to students.json!")
```

```python
#Read data from a JSON file.

with open("students.json", "r") as file:
    students = json.load(file)
    
print("Students loaded from file:")
for student in students:
    print(f"Name: {student['name']}, Age: {student['age']}, Grade: {student['grade']}")
```

```python
#Update an existing student's information.

with open("students.json", "r") as file:
    students = json.load(file)

name_to_update = "Bilal"
for student in students:
    if student["name"] == name_to_update:
        student["grade"] = "A"
        student["age"] = 23
        print(f"{name_to_update}'s record updated!")
        break
else:
    print(f"Student '{name_to_update}' not found. ")

with open("students.json", "w") as file:
    json.dump(students, file, indent=4)
```

```python
#Add a new student to the JSON file.

with open("students.json", "r") as file:
    students = json.load(file)

new_student = {"name": "Malaika", "age": 19, "grade": "B+"}
students.append(new_student)

with open("students.json", "w") as file:
    json.dump(students, file, indent=4)

print(f"New student '{new_student['name']}' added successfully!")
```
