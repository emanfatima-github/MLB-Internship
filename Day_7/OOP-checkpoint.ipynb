```python
# Student Class
```

```python
class Student:

    def __init__(self, name, age, roll_no):
        self.name = name
        self.age = age
        self.roll_no = roll_no

    def display(self):
        print("Student Information")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Roll No:", self.roll_no)


student1 = Student("Eman", 21, "11")
student1.display()
```

```python
# Employee Class
```

```python
class Employee:

    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.salary = salary

    def display(self):
        print("\nEmployee Information")
        print("Name:", self.name)
        print("Department:", self.department)
        print("Salary:", self.salary)


emp1 = Employee("Ali", "IT", 50000)
emp1.display()
```

```python
# Car Class
```

```python
class Car:

    def __init__(self, brand, model, color):
        self.brand = brand
        self.model = model
        self.color = color

    def start(self):
        print(self.brand,"is starting.")

    def stop(self):
        print(self.brand,"has stopped.")

    def display(self):
        print("\nCar Details")
        print("Brand:",self.brand)
        print("Model:",self.model)
        print("Color:",self.color)


car1 = Car("Toyota","Corolla","White")
car1.display()
car1.start()
car1.stop()
```
