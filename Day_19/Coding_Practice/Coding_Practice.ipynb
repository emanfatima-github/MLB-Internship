```python
import cv2
import numpy as np
```

```python
image = cv2.imread("document.jpg")

if image is None:
    print("Image not found!")
else:
    print("Image Loaded Successfully")
```

```python
while True:

    print("\n========== IMAGE PROCESSING MENU ==========")
    print("1. Translate Image")
    print("2. Rotate Image")
    print("3. Scale Image")
    print("4. Affine Transformation")
    print("5. Perspective Transformation")
    print("6. Increase Brightness")
    print("7. Decrease Brightness")
    print("8. Increase Contrast")
    print("9. Decrease Contrast")
    print("10. Gaussian Blur")
    print("11. Median Blur")
    print("12. Bilateral Filter")
    print("13. Sharpen Image")
    print("14. Exit")

    choice = int(input("Enter your choice: "))

    img = image.copy()

    if choice == 1:
        tx = int(input("Move Horizontally (pixels): "))
        ty = int(input("Move Vertically (pixels): "))

        M = np.float32([[1,0,tx],
                        [0,1,ty]])

        result = cv2.warpAffine(img,M,(img.shape[1],img.shape[0]))

        cv2.imshow("Translated",result)
        cv2.imwrite("output/translated.jpg",result)

    elif choice == 2:
        angle = float(input("Enter Rotation Angle: "))

        center = (img.shape[1]//2,img.shape[0]//2)

        M = cv2.getRotationMatrix2D(center,angle,1)

        result = cv2.warpAffine(img,M,(img.shape[1],img.shape[0]))

        cv2.imshow("Rotated",result)
        cv2.imwrite("output/rotated.jpg",result)

    elif choice == 3:
        scale = float(input("Enter Scaling Factor (0.5 or 2): "))

        result = cv2.resize(img,None,fx=scale,fy=scale)

        cv2.imshow("Scaled",result)
        cv2.imwrite("output/scaled.jpg",result)

    elif choice == 4:
        rows,cols = img.shape[:2]

        pts1 = np.float32([[50,50],[200,50],[50,200]])

        pts2 = np.float32([[10,100],[200,50],[100,250]])

        M = cv2.getAffineTransform(pts1,pts2)

        result = cv2.warpAffine(img,M,(cols,rows))

        cv2.imshow("Affine",result)
        cv2.imwrite("output/affine.jpg",result)

    elif choice == 5:
        rows,cols = img.shape[:2]

        pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])

        pts2 = np.float32([[0,0],[300,0],[0,400],[300,400]])

        M = cv2.getPerspectiveTransform(pts1,pts2)

        result = cv2.warpPerspective(img,M,(300,400))

        cv2.imshow("Perspective",result)
        cv2.imwrite("output/perspective.jpg",result)

    elif choice == 6:
        result = cv2.convertScaleAbs(img,alpha=1,beta=50)

        cv2.imshow("Brightness Increased",result)
        cv2.imwrite("output/brightness_up.jpg",result)

    elif choice == 7:
        result = cv2.convertScaleAbs(img,alpha=1,beta=-50)

        cv2.imshow("Brightness Decreased",result)
        cv2.imwrite("output/brightness_down.jpg",result)

    elif choice == 8:
        result = cv2.convertScaleAbs(img,alpha=1.5,beta=0)

        cv2.imshow("Contrast Increased",result)
        cv2.imwrite("output/contrast_up.jpg",result)

    elif choice == 9:
        result = cv2.convertScaleAbs(img,alpha=0.6,beta=0)

        cv2.imshow("Contrast Decreased",result)
        cv2.imwrite("output/contrast_down.jpg",result)

    elif choice == 10:
        result = cv2.GaussianBlur(img,(9,9),0)

        cv2.imshow("Gaussian Blur",result)
        cv2.imwrite("output/gaussian.jpg",result)

    elif choice == 11:
        result = cv2.medianBlur(img,5)

        cv2.imshow("Median Blur",result)
        cv2.imwrite("output/median.jpg",result)

    elif choice == 12:
        result = cv2.bilateralFilter(img,9,75,75)

        cv2.imshow("Bilateral",result)
        cv2.imwrite("output/bilateral.jpg",result)

    elif choice == 13:
        kernel = np.array([[0,-1,0],
                           [-1,5,-1],
                           [0,-1,0]])

        result = cv2.filter2D(img,-1,kernel)

        cv2.imshow("Sharpened",result)
        cv2.imwrite("output/sharpen.jpg",result)

    elif choice == 14:
        print("Program Ended")
        break

    else:
        print("Invalid Choice")

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
