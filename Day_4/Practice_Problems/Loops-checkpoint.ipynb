```python
n = int(input("Enter a number: "))
sum = 0
for i in range(1,n+1):
    sum = sum + i

print("Sum: ",sum)
```

```python
num = int(input("Enter a number: "))
for i in range(1,11):
    print(num, "x", i, "=", num*i)
```

```python
num = int(input("Enter a number: "))
count = 0

while num > 0:
    count = count + 1
    num = num // 10

print("Digits =", count)
```
