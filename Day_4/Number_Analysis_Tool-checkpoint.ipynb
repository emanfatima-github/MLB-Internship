```python
# Number Analysis Tool

num = int(input("Enter a number: "))

# Even or Odd
if num % 2 == 0:
    even_odd = "Even"
else:
    even_odd = "Odd"

# Prime Check
count = 0
for i in range(1, num + 1):
    if num % i == 0:
        count += 1

if count == 2:
    prime = "Yes"
else:
    prime = "No"

# Count Digits
temp = num
digits = 0

if temp == 0:
    digits = 1
else:
    while temp > 0:
        digits += 1
        temp = temp // 10

# Reverse Number
temp = num
reverse = 0

while temp > 0:
    digit = temp % 10
    reverse = reverse * 10 + digit
    temp = temp // 10

# Palindrome Check
if num == reverse:
    palindrome = "Yes"
else:
    palindrome = "No"

# Display Results
print("\n----- Number Analysis Report -----")
print("Number        :", num)
print("Even/Odd      :", even_odd)
print("Prime         :", prime)
print("Digit Count   :", digits)
print("Reverse       :", reverse)
print("Palindrome    :", palindrome)
```
