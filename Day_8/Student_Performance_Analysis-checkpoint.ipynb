```python
import pandas as pd

# Load the dataset
df = pd.read_csv("student_performance.csv")

# Display basic information
print("========== FIRST 5 ROWS ==========")
print(df.head())

print("\n========== DATASET INFORMATION ==========")
df.info()

print("\n========== SUMMARY STATISTICS ==========")
print(df.describe())

# Calculate average marks for each subject
print("\n========== AVERAGE MARKS FOR EACH SUBJECT ==========")
print(f"\tPython             : {df['Python'].mean():.2f}")
print(f"\tMathematics        : {df['Mathematics'].mean():.2f}")
print(f"\tStatistics         : {df['Statistics'].mean():.2f}")
print(f"\tMachine Learning   : {df['Machine_Learning'].mean():.2f}")

# Calculate total marks of each student
df["Total"] = (
    df["Python"]
    + df["Mathematics"]
    + df["Statistics"]
    + df["Machine_Learning"]
)

# Identify the top 5 performing students
print("\n========== TOP 5 PERFORMING STUDENTS ==========")
top5 = df.sort_values(by="Total", ascending=False)
print(top5[["Student_ID", "Name", "Total"]].head().to_string(index=False))

# Find average total marks
average_total = df["Total"].mean()

# Find students scoring below the average
print("\n========== STUDENTS SCORING BELOW AVERAGE ==========")
below_average = df[df["Total"] < average_total]
print(below_average[["Student_ID", "Name", "Total"]].to_string(index=False))

# Display the total number of students
print("\n========== TOTAL NUMBER OF STUDENTS ==========")
print(f"\tTotal Students : {len(df)}")

# Save the analyzed dataset
df.to_csv("student_performance_analysis.csv", index=False)

print("\n========== FILE SAVED ==========")
print("\tAnalysis saved successfully as 'student_performance_analysis.csv'")
```
