```python
# Load the dataset using Pandas.
```

```python
import pandas as pd

df = pd.read_csv("students.csv")
print(df)
```

```python
#  Display basic information about the dataset
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
# average marks for each subject
```

```python
print("Average Math Marks:", df["Math"].mean())
print("Average English Marks:", df["English"].mean())
print("Average Science Marks:", df["Science"].mean())
```

```python
# Identify the Top 5 Performing Students
```

```python
df["Total"] = df["Math"] + df["English"] + df["Science"]

top5 = df.sort_values(by="Total", ascending=False)
print(top5.head())
```

```python
# Find Students Scoring Below the Average
```

```python
average = df["Total"].mean()

below_average = df[df["Total"] < average]
print(below_average)
```

```python
# Display the Total Number of Studen
```

```python
print("Total Students:", len(df))
```

```python
# Save the Analyzed Dataset as a New CSV File
```

```python
df.to_csv("students_analysis.csv", index=False)
print("New CSV file saved successfully.")
```
