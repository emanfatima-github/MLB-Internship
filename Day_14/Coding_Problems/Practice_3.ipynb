```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# ReLU Model
```

```python
print("ReLU Model")
relu_model = keras.Sequential([
    layers.Input(shape=(4,)),
    layers.Dense(8, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])
relu_model.summary()
```

```python
# Sigmoid Model
```

```python
print("\nSigmoid Model")
sigmoid_model = keras.Sequential([
    layers.Input(shape=(4,)),
    layers.Dense(8, activation="sigmoid"),
    layers.Dense(1, activation="sigmoid")
])
sigmoid_model.summary()
```

```python
# Tanh Model
```

```python
print("\nTanh Model")
tanh_model = keras.Sequential([
    layers.Input(shape=(4,)),
    layers.Dense(8, activation="tanh"),
    layers.Dense(1, activation="sigmoid")
])
tanh_model.summary()
```

```python
#Changing the activation function (ReLU, Sigmoid, or Tanh) does not affect the model structure.
#It only changes how the neurons process the input data and learn patterns, which can influence
#the model's performance during training.
```
