```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_student_performance.csv")   
df.head()
```

```python
# Bar chart showing the average marks of each student.
```

```python
plt.figure(figsize=(10,5))
plt.bar(df["Student_Name"], df["Average_Score"])

plt.title("Average Marks of Each Student")
plt.xlabel("Students")
plt.ylabel("Average Score")
plt.xticks(rotation=45)

plt.show()
```

```python
# Histogram showing the distribution of Average Scores.
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Average_Score"], bins=10)

plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")

plt.show()
```

```python
# Scatter plot comparing Python and Machine Learning marks.
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Python"], df["Machine_Learning"])

plt.title("Python vs Machine Learning Marks")
plt.xlabel("Python Marks")
plt.ylabel("Machine Learning Marks")

plt.show()
```

```python
# Pie chart showing the distribution of students by Performance category.
```

```python
performance_counts = df["Performance"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(performance_counts,
        labels=performance_counts.index,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Distribution of Students by Performance")

plt.show()
```

```python
# Box plot to visualize the spread of marks in all subjects.
```

```python
subject_columns = ["Python", "Mathematics", "Statistics", "Machine_Learning"]

plt.figure(figsize=(8,5))
plt.boxplot(df[subject_columns])

plt.title("Spread of Marks in All Subjects")
plt.xticks(range(1, len(subject_columns)+1), subject_columns)
plt.ylabel("Marks")

plt.show()
```
