```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student_performance.csv")
```

```python
# Check for missing values
```

```python
print("Missing Values: ")
print(df.isnull().sum())
```

```python
# Remove duplicate records
```

```python
df = df.drop_duplicates()
```

```python
# Rename columns
```

```python
df = df.rename(columns = {
    "Name" : "Student_Name",
    "Age" : "Student_Age"
})
```

```python
# Create Average Score Column
```

```python
subject_columns = ["Python", "Mathematics", "Statistics", "Machine_Learning"]

df["Average_Score"] = df[subject_columns].mean(axis=1)
```

```python
# Create Performance Column
```

```python
def performance(score):
    if score >= 90:
        return "Excellent"
    elif score >= 80:
        return "Good"
    elif score >= 70:
        return "Average"
    else:
        return "Needs Improvement"

df["Performance"] = df["Average_Score"].apply(performance)

# display first 5 rows
print("\nCleaned Dataset: ")
print(df.head())

#Save the cleaned dataset
df.to_csv("cleaned_student_performance.csv", index=False)

print("\nCleaned dataset saved as 'cleaned_student_performance.csv'")
```
